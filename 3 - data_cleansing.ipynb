```python
# read the csv file
import pandas as pd
my_df = pd.read_csv('Stage 2.3 - new papers full text without ref.csv')
#del my_df['Unnamed: 0']
#del my_df['english lang']
#del my_df['blank']
#del my_df['remove manually']
del my_df['index']
print(my_df.shape)


```

```python
# remove new line characters
import re

for i in my_df.index:
    my_df['Full text without references'][i] = re.sub('\s+', ' ', my_df['Full text without references'][i])
print(my_df['Full text without references'][916])
```

```python
# remove email addresses
import re

for i in my_df.index:
    my_df['Full text without references'][i] = re.sub('^([a-zA-Z0-9_\-\.\s]+)@([a-zA-Z0-9_\-\.]+)\.([a-zA-Z]{2,5})$|(?:\.?)([\w\-_+#~!$&\'\.]+(?<!\.)(@|[ ]?\(?[ ]?(at|AT)[ ]?\)?[ ]?)(?<!\.)[\w]+[\w\-\.]*\.[a-zA-Z-]{2,3})(?:[^\w])', '', my_df['Full text without references'][i])
print(my_df['Full text without references'][916])
```

```python
# remove website addresses
import re

for i in my_df.index:
    my_df['Full text without references'][i] = re.sub('(https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|www\.[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9]+\.[^\s]{2,}|www\.[a-zA-Z0-9]+\.[^\s]{2,})', '', my_df['Full text without references'][i])
print(my_df['Full text without references'][916])
```

```python
# remove repetitive text
import re

for i in my_df.index:
    my_df['Full text without references'][i] = re.sub('''Reproduced with permission of copyright owner.|IEEE|UTC from  Xplore|Authorized licensed use limited to|Ryerson University Library|Restrictions apply.|Further reproduction prohibited without permission|Authorized licensed use limited to: Ryerson University Library.|Downloaded on|from IEEE Xplore.''', '', my_df['Full text without references'][i])
print(my_df['Full text without references'][916])
    # remove: Reproduced with permission of copyright owner.
#  Further reproduction prohibited without permission
# Authorized licensed use limited to: Ryerson University Library. Downloaded on May 03,2022 at 02:31:36 UTC from IEEE Xplore.  Restrictions apply. 

```

```python
# remove date formats
import re

for i in my_df.index:
    my_df['Full text without references'][i] = re.sub('(\d{2}[ /-](?:\d{2}|Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)[ /-]\d{4})', '', my_df['Full text without references'][i])
print(my_df['Full text without references'][916])
    
 
```

```python
# remove numbers
import re


for i in my_df.index:
    my_df['Full text without references'][i] = re.sub('\d+', '', my_df['Full text without references'][i])
print(my_df['Full text without references'][60])
```

```python
# remove unnecessary text
import re

# consider them as stop words - add them into the stop words
# analysis on how we can update the stopwords to see how the outcome changes

for i in my_df.index:
    my_df['Full text without references'][i] = re.sub('(cid: )| et al.|et al|et al|cid_cid|(cid:cid)', '', my_df['Full text without references'][i])
print(my_df['Full text without references'][916])
```

```python
# export to csv
my_df.to_csv('Stage 3 - Relevant new papers.csv', index = False)    
                                                # this is the final CSV file containing full-text relevant new papers
```
