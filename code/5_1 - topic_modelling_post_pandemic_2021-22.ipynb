```python
# splitting the sentences into a list of words using gensim’s simple_preprocess() function
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc = True, min_len=3))   # deacc = True removes punctuations, min_len = 3 only considers words of len >= 3

data_words = list(sent_to_words(data))

print(data_words[ :1])
```

```python
# building the bigram and trigram models (# only those that appear min_count times or more; higher threshold fewer phrases)
bigram = gensim.models.Phrases(data_words, min_count = 40, threshold = 10)     # min_count set to 5% of the number of documents
trigram = gensim.models.Phrases(bigram[data_words], threshold = 10)  

bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

# reporting a trigram example
print(trigram_mod[bigram_mod[data_words[0]]])
```

```python
# defining functions for removing stopwords, building bigrams and trigrams, lemmatization, and excluding other parts of speech
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags = ['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
## running the above functions

# removing stop Words 
data_words_nostops = remove_stopwords(data_words)

# creating bigrams
data_words_bigrams = make_bigrams(data_words_nostops)

# initializing spacy 'en' model, keeping only tagger component for efficiency
# if it doesn't load, run this in anaconda prompt as administrator: python -m spacy download en
nlp = spacy.load('en_core_web_sm', disable = ['parser', 'ner'])

# lemmatizing and keeping only nouns and verbs ('ADJ', and 'ADV' removed)
data_lemmatized = lemmatization(data_words_bigrams, allowed_postags = ['NOUN', 'VERB'])
data_lemmatized = remove_stopwords(data_lemmatized)
print(data_lemmatized[ :1])
#data_lemmatized = remove_stopwords(data_words)
```

```python
# to get the most common bigrams
import collections
cnt = collections.Counter()

for i in data_lemmatized:
    cnt.update(nltk.bigrams(i))

print(cnt.most_common(25))
```

```python
# creating dictionary - here is the point: data_lemmatized is a list of words in each document
# for our purpose, we need to create a second data_lemmatized based on our keywords - it is a list of strings
id2word = corpora.Dictionary(data_lemmatized)

# creating corpus
texts = data_lemmatized

# term document frequency
corpus = [id2word.doc2bow(text) for text in texts]

# report the results
pprint(corpus[:1])
```

```python
# determining what topic each document is about
data_years = my_df.Year.values
data_codes = my_df.Code.values
def format_topics_sentences(ldamodel = new_model, corpus = corpus, texts = data, years = data_years, codes = data_codes):
    # init output
    sent_topics_df = pd.DataFrame()

    # getting main topic in each document
    for i, row in enumerate(ldamodel[corpus]):
        row = sorted(row, key = lambda x: (x[1]), reverse = True)
        # getting the dominant topic, percentage of contribution and keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num, topn=20)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic, 4), topic_keywords]), ignore_index = True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # adding original text to the end of the output
    contents = pd.Series(texts)
    content_years = pd.Series(years)
    content_codes = pd.Series(codes)
    sent_topics_df = pd.concat([sent_topics_df, contents,content_years,content_codes], axis = 1)
    return(sent_topics_df)


df_topic_sents_keywords = format_topics_sentences(ldamodel = new_model, corpus = corpus, texts = data, years = data_years, codes = data_codes)

# formatting
df_dominant_topic = df_topic_sents_keywords.reset_index()
df_dominant_topic.columns = ['Document_No', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'Text','Year','Code']

# reporting
df_dominant_topic.tail(10)
```
